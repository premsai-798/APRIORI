```python
import random
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
from itertools import chain, combinations

# Define the list of items
items = ['Milk', 'Bread', 'Eggs', 'Butter', 'Cheese', 'Jam', 'Coffee', 'Sugar', 'Tea', 'Biscuits']

# Generate random transactions
num_transactions = 1000
transactions = [set(random.sample(items, random.randint(2, 5))) for _ in range(num_transactions)]

# Displaying a few transactions
print("Sample Transactions:")
for i in range(5):
    print(transactions[i])

# Function to calculate support of an itemset

def calculate_support(itemset, transactions):
    count = 0
    for transaction in transactions:
        if itemset.issubset(transaction):
            count += 1
    return count / len(transactions)

# Generate frequent itemsets
min_support = 0.01
frequent_itemsets = {}

# Generating 1-itemsets
itemsets = defaultdict(int)
for transaction in transactions:
    for item in transaction:
        itemsets[frozenset([item])] += 1

num_transactions = len(transactions)
frequent_itemsets = {itemset: count / num_transactions for itemset, count in itemsets.items() if (count / num_transactions) >= min_support}

# Convert frequent itemsets to DataFrame for plotting
import pandas as pd

top_itemsets_df = pd.DataFrame(list(frequent_itemsets.items()), columns=['Itemset', 'Support'])
top_itemsets_df = top_itemsets_df.sort_values(by='Support', ascending=False).head(10)
```
